### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
suma_total_precio = tienda['Precio'].sum()
suma_total_precio2 = tienda2['Precio'].sum()
suma_total_precio3 = tienda3['Precio'].sum()
suma_total_precio4 = tienda4['Precio'].sum()

```

```python
print(f'El valor total de las ventas de la tienda 1 es: ${suma_total_precio}')
print(f'El valor total de las ventas de la tienda 2 es: ${suma_total_precio2}')
print(f'El valor total de las ventas de la tienda 3 es: ${suma_total_precio3}')
print(f'El valor total de las ventas de la tienda 4 es: ${suma_total_precio4}')
```

```python
print(f'Las ventas por categoría de la tienda 1 son: \n{ventas_por_categoria}')
print(f'\nLas ventas por categoría de la tienda 2 son: \n{ventas_por_categoria_2}')
print(f'\nLas ventas por categoría de la tienda 3 son: \n{ventas_por_categoria_3}')
print(f'\nLas ventas por categoría de la tienda 4 son: \n{ventas_por_categoria_4}')
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
# ---- Productos más y menos vendidos por tienda ----

def resumen_productos(df, nombre_tienda):
    conteo = df['Producto'].value_counts()

    mas_vendido = conteo.idxmax()
    ventas_mas = conteo.max()

    menos_vendido = conteo.idxmin()
    ventas_menos = conteo.min()

    print(f"\n===== {nombre_tienda} =====")
    print(f"Producto más vendido:  {mas_vendido}  ({ventas_mas} ventas)")
    print(f"Producto menos vendido: {menos_vendido}  ({ventas_menos} ventas)")
    print("\nTop 5 más vendidos:")
    print(conteo.head())
    print("\nTop 5 menos vendidos:")
    print(conteo.tail())


# ---- Ejecutar para cada tienda ----

resumen_productos(tienda,  "Tienda 1")
resumen_productos(tienda2, "Tienda 2")
resumen_productos(tienda3, "Tienda 3")
resumen_productos(tienda4, "Tienda 4")

```

```python
# ---- Productos que generaron más y menos dinero por tienda ----

def resumen_ingresos(df, nombre_tienda):
    ingresos = df.groupby('Producto')['Precio'].sum().sort_values(ascending=False)

    prod_mas = ingresos.idxmax()
    dinero_mas = ingresos.max()

    prod_menos = ingresos.idxmin()
    dinero_menos = ingresos.min()

    print(f"\n===== {nombre_tienda} =====")
    print(f"Producto que generó MÁS dinero:    {prod_mas}  (${dinero_mas:,.0f})")
    print(f"Producto que generó MENOS dinero:  {prod_menos} (${dinero_menos:,.0f})")

    print("\nTop 5 productos que más dinero generaron:")
    print(ingresos.head())

    print("\nTop 5 productos que menos dinero generaron:")
    print(ingresos.tail())


# ---- Ejecutar para cada tienda ----

resumen_ingresos(tienda,  "Tienda 1")
resumen_ingresos(tienda2, "Tienda 2")
resumen_ingresos(tienda3, "Tienda 3")
resumen_ingresos(tienda4, "Tienda 4")
```

```python
# ---- Productos que generaron más y menos dinero + cantidad vendida ----

def resumen_ingresos(df, nombre_tienda):

    # Cantidad de ventas por producto
    cantidad = df['Producto'].value_counts()

    # Dinero generado por producto
    ingresos = df.groupby('Producto')['Precio'].sum()

    # Unir todo en una sola tabla
    tabla = pd.concat([cantidad, ingresos], axis=1)
    tabla.columns = ['Cantidad Vendida', 'Total Ganado']
    tabla = tabla.sort_values('Total Ganado', ascending=False)

    # Identificar top y bottom
    prod_mas = tabla.index[0]
    dinero_mas = tabla.iloc[0, 1]
    cant_mas = tabla.iloc[0, 0]

    prod_menos = tabla.index[-1]
    dinero_menos = tabla.iloc[-1, 1]
    cant_menos = tabla.iloc[-1, 0]

    print(f"\n===== {nombre_tienda} =====")
    print(f"Producto que generó MÁS dinero: {prod_mas}")
    print(f"  → Total ganado: ${dinero_mas:,.0f}")
    print(f"  → Cantidad vendida: {cant_mas}")

    print(f"\nProducto que generó MENOS dinero: {prod_menos}")
    print(f"  → Total ganado: ${dinero_menos:,.0f}")
    print(f"  → Cantidad vendida: {cant_menos}")

    print("\nTop 5 productos con mayor ganancia:")
    print(tabla.head())

    print("\nTop 5 productos con menor ganancia:")
    print(tabla.tail())


# ---- Ejecutar para cada tienda ----

resumen_ingresos(tienda,  "Tienda 1")
resumen_ingresos(tienda2, "Tienda 2")
resumen_ingresos(tienda3, "Tienda 3")
resumen_ingresos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
def costo_envio_promedio(df, nombre):
    promedio = df['Costo de envío'].mean()
    print(f"{nombre}: ${promedio:.2f}")

print("Costo de envío promedio por tienda:\n")
costo_envio_promedio(tienda, "Tienda 1")
costo_envio_promedio(tienda2, "Tienda 2")
costo_envio_promedio(tienda3, "Tienda 3")
costo_envio_promedio(tienda4, "Tienda 4")
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("ggplot")

# ============ 1. Ingresos totales por tienda ============

ingresos_totales = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

max_t = max(ingresos_totales, key=ingresos_totales.get)
min_t = min(ingresos_totales, key=ingresos_totales.get)

colores = [
    "#2ecc71" if t == max_t else
    "#e74c3c" if t == min_t else
    "#3498db"
    for t in ingresos_totales
]

plt.figure(figsize=(10,6))
plt.bar(ingresos_totales.keys(), ingresos_totales.values(), color=colores, edgecolor="black")
plt.title("Ingresos Totales por Tienda", fontsize=16)
plt.xlabel("Tiendas", fontsize=14)
plt.ylabel("Ingresos ($)", fontsize=14)
plt.grid(alpha=0.2)
plt.show()


# ============ 2. Ventas vs Ganancias por tienda  ============

ventas = [
    len(tienda),
    len(tienda2),
    len(tienda3),
    len(tienda4)
]

ganancias = list(ingresos_totales.values())
labels = list(ingresos_totales.keys())

fig, ax1 = plt.subplots(figsize=(12,6))

# Barras en eje izquierdo
ax1.bar(labels, ventas, color="#5DADE2", edgecolor="black", width=0.6)
ax1.set_ylabel("Cantidad de Ventas", fontsize=14)
ax1.set_xlabel("Tiendas", fontsize=14)
ax1.tick_params(axis='y')
ax1.set_title("Relación entre Cantidad de Ventas y Ganancias por Tienda", fontsize=16)

# Línea en eje derecho
ax2 = ax1.twinx()
ax2.plot(labels, ganancias, marker="o", markersize=10, linewidth=3, color="#8E44AD")
ax2.set_ylabel("Ganancias Totales ($)", fontsize=14)
ax2.tick_params(axis='y')

plt.grid(alpha=0.2)
plt.show()



# ============ 3. Ingreso promedio por venta ============

ingreso_promedio = {
    "T1": tienda["Precio"].mean(),
    "T2": tienda2["Precio"].mean(),
    "T3": tienda3["Precio"].mean(),
    "T4": tienda4["Precio"].mean()
}

plt.figure(figsize=(9,6))
plt.bar(ingreso_promedio.keys(), ingreso_promedio.values(),
        color="#F4D03F", edgecolor="black")

plt.title("Ingreso Promedio por Venta", fontsize=16)
plt.xlabel("Tienda", fontsize=14)
plt.ylabel("Ingreso ($)", fontsize=14)
plt.grid(alpha=0.25)
plt.show()


# ============ 4. Ventas y Costos Totales por Categoría ============


union = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

ventas_cat = union["Categoría del Producto"].value_counts()
costos_cat = union.groupby("Categoría del Producto")["Precio"].sum()

fig, ax1 = plt.subplots(figsize=(14,7))

# Barras para cantidad de ventas
ax1.bar(
    ventas_cat.index,
    ventas_cat.values,
    color="#48C9B0",
    edgecolor="black",
    alpha=0.85,
    label="Ventas por categoría"
)
ax1.set_ylabel("Cantidad de Ventas", fontsize=14)
ax1.set_xlabel("Categoría del Producto", fontsize=14)
ax1.tick_params(axis='y')
ax1.set_xticklabels(ventas_cat.index, rotation=45)

ax2 = ax1.twinx()
ax2.scatter(
    costos_cat.index,
    costos_cat.values,
    color="#C0392B",
    s=120,         # tamaño del punto
    marker="o",
    label="Costo total por categoría"
)
ax2.set_ylabel("Costo Total ($)", fontsize=14)
ax2.tick_params(axis='y')

plt.title("Ventas y Costos Totales por Categoría (Barras + Puntos)", fontsize=16)
fig.legend(loc="upper right", fontsize=12)

plt.grid(alpha=0.2)
plt.show()




# ============ 5. Ingreso por Tienda según el Tiempo ============

# Conversión de fechas
tienda["Fecha"] = pd.to_datetime(tienda["Fecha de Compra"], dayfirst=True)
tienda2["Fecha"] = pd.to_datetime(tienda2["Fecha de Compra"], dayfirst=True)
tienda3["Fecha"] = pd.to_datetime(tienda3["Fecha de Compra"], dayfirst=True)
tienda4["Fecha"] = pd.to_datetime(tienda4["Fecha de Compra"], dayfirst=True)

# Agrupación por mes (resample necesita índice datetime)
t1 = tienda.set_index("Fecha").resample("ME")["Precio"].sum()
t2 = tienda2.set_index("Fecha").resample("ME")["Precio"].sum()
t3 = tienda3.set_index("Fecha").resample("ME")["Precio"].sum()
t4 = tienda4.set_index("Fecha").resample("ME")["Precio"].sum()

plt.figure(figsize=(14,7))

plt.plot(t1.index, t1.values, marker="o", linewidth=2.5, label="Tienda 1")
plt.plot(t2.index, t2.values, marker="o", linewidth=2.5, label="Tienda 2")
plt.plot(t3.index, t3.values, marker="o", linewidth=2.5, label="Tienda 3")
plt.plot(t4.index, t4.values, marker="o", linewidth=2.5, label="Tienda 4")

plt.title("Ingresos por Tienda a lo Largo del Tiempo", fontsize=16)
plt.xlabel("Fecha", fontsize=14)
plt.ylabel("Ingresos Mensuales ($)", fontsize=14)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)

plt.show()
```
